```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import warnings

warnings.simplefilter('ignore')


#Set up original data
data = pd.read_csv("data.csv")
print("Length of DataFrame before subsample:", len(data))


#Reduce the size of dataset to 1000
original_data = data.sample(n=50000).reset_index(drop = True)
original_data.head()
print("Length of DataFrame after subsample:", len(original_data))
```

**1: Cleaning data**

```python
#Cleaning the original data
#Check missing values and duplication
original_data.isnull().sum()

#Check Duplication
#original_data["duplication"]=original_data.duplicated()
no_duplicated_data=original_data.drop_duplicates()

#Adding Recency columns
no_duplicated_data['Delivery_Date'] = pd.to_datetime(no_duplicated_data['Delivery_Date'],format='%d%b%Y')
no_duplicated_data['Recency'] = (pd.to_datetime('today') - pd.to_datetime(no_duplicated_data['Delivery_Date'])).dt.days

#Adding Age columns
no_duplicated_data[ 'DOB'] = pd.to_datetime(no_duplicated_data[ 'Customer_BirthDate' ], format= '%d%b%Y')
today = pd.Timestamp( 'now')
no_duplicated_data[ 'Age'] = (today - no_duplicated_data[ 'DOB']).dt.days//365
#print(no_duplicated_data)

#Adding CLV
no_duplicated_data[ 'Total Revenue' ] = no_duplicated_data[ 'Total Revenue' ].apply(lambda x: x.replace('$', '')).apply(lambda x: x.replace(',', ''))
no_duplicated_data.rename(columns={'Total Revenue': "CLV"}, inplace=True)
no_duplicated_data["CLV"]=no_duplicated_data["CLV"].astype(float)

#Adding Avg_Unit_Cost
no_duplicated_data[ 'Unit Cost' ] = no_duplicated_data[ 'Unit Cost' ].apply(lambda x: x.replace('$', '')).apply(lambda x: x.replace(',', ''))
no_duplicated_data.rename(columns={'Unit Cost': "Avg_Unit_Cost"}, inplace=True)
no_duplicated_data["Avg_Unit_Cost"]=no_duplicated_data["Avg_Unit_Cost"].astype(float)

#print(no_duplicated_data.dtypes)
print(no_duplicated_data)

```

```python
#Check outliers


no_duplicated_data["CLV"]=no_duplicated_data["CLV"].apply(lambda x: round(x,2))

def check_outliers(data, column):
    sorts = data[column].sort_values()
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print("Lower bound: "+str(lower_bound))
    print("Upper bound: "+str(upper_bound))

    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data[column])
    plt.title(f'Boxplot of {column}')
    plt.show()

#Deploy each feature
check_outliers(no_duplicated_data, 'Recency')
check_outliers(no_duplicated_data, 'CLV')
check_outliers(no_duplicated_data, 'Avg_Unit_Cost')
check_outliers(no_duplicated_data, 'Age')

#Check upperbound for CLV outliers
sorts = no_duplicated_data["CLV"].sort_values()
Q1C = no_duplicated_data["CLV"].quantile(0.25)
Q3C = no_duplicated_data["CLV"].quantile(0.75)
IQRC = Q3C - Q1C
lower_boundC = Q1C - 1.5 * IQRC
upper_boundC = Q3C + 1.5 * IQRC

#check upperbound for Avg_Unit_Cost outliers
sorts = no_duplicated_data["Avg_Unit_Cost"].sort_values()
Q1D = no_duplicated_data["Avg_Unit_Cost"].quantile(0.25)
Q3D = no_duplicated_data["Avg_Unit_Cost"].quantile(0.75)
IQRD = Q3D - Q1D
lower_boundD = Q1D - 1.5 * IQRD
upper_boundD = Q3D + 1.5 * IQRD

#Remove outliers
no_duplicated_data = no_duplicated_data[(no_duplicated_data["CLV"]<upper_boundC) & (no_duplicated_data["Avg_Unit_Cost"]<upper_boundD)]

```

```python
#Grouping by Customer
#Recency: I pick the very recent shipping date
#CLV: Total amount that the customer spent
#Avg_Unit_Cost: the average price for all price
data_by_customerID = no_duplicated_data.groupby( 'Customer ID' ).agg( {"Order ID":"count","Recency":"min","CLV":"sum", "Avg_Unit_Cost":"mean","Age":"min"}).reset_index()
#result = no_duplicated_data.groupby('group').agg(sum_value1=('value1', 'sum'), mean_value2=('value2', 'mean'))({"sess_length": [ 'sum', 'count','mean']})
data_by_customerID.rename(columns={'Order ID': "Frequency"}, inplace=True)
print(data_by_customerID[data_by_customerID["Frequency"]>1])
```

```python
#import library
from sklearn.metrics import silhouette_score, silhouette_samples
import matplotlib.cm as cm

X_sil=pca_cust_data.iloc[:,0:2]
print(X_sil)
Y_sil=pca_cust_data.iloc[:,-1]
print(Y_sil)

#Silhouette method
for n_clusters in range(2,7):
    # Create a subplot with 1 row and 2 columns.
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot.
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1].
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X_sil) + (n_clusters + 1) * 10])
    # Initialize the clusterer_sil with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer_sil = KMeans(n_clusters=n_clusters, random_state=10, n_init = 10)
    cluster_label = clusterer_sil.fit_predict(X_sil)
    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    sil_coeff = silhouette_score(X_sil, cluster_label)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", sil_coeff)
    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X_sil, cluster_label)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them.
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_label == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle.
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot.
        y_lower = y_upper + 10  # 10 for the 0 samples.
    # Labeling the clusters
    centers = clusterer_sil.cluster_centers_
    # Draw white circles at cluster centers.
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o', c="white", alpha=1, s=200, edgecolor='k')

    #to numpy
    X_sil_arr=X_sil.to_numpy()
    # Plot the silhouette values
    ax2.scatter(X_sil_arr[:, 0], X_sil_arr[:, 1], marker='.', s=200, lw=0, alpha=0.7, c=cluster_label)


    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1, s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters), fontsize=14, fontweight='bold')

plt.show()
```
